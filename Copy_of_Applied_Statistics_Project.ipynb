# Step 4: Correct data types

```python
# Convert Dt_Customer to datetime with DD-MM-YYYY HH:MM format
def convert_excel_date(value):
    try:
        # If value is numeric (Excel serial date)
        if isinstance(value, (int, float)):
            return pd.to_datetime('1899-12-30') + pd.to_timedelta(value, unit='D')
        # If value is string in MM/DD/YYYY format
        return pd.to_datetime(value, format='%m/%d/%Y', errors='coerce')
    except:
        return pd.NaT

df['Dt_Customer'] = df['Dt_Customer'].apply(convert_excel_date)
nat_count = df['Dt_Customer'].isna().sum()
print(f"\nNumber of NaT values after conversion: {nat_count}")

# Impute NaT with median of valid dates or fallback
valid_dates = df['Dt_Customer'].dropna()
if not valid_dates.empty:
    median_date = valid_dates.median()
    df['Dt_Customer'] = df['Dt_Customer'].fillna(median_date)
    print(f"Imputed {nat_count} NaT values with median date: {median_date}")
else:
    default_date = pd.to_datetime('2023-01-01')
    df['Dt_Customer'] = df['Dt_Customer'].fillna(default_date)
    print(f"All dates invalid; imputed with: {default_date}")
```

Converted to datetime data type and filled those missing NaT entries with the median value to fill them for consistency.

```python
# Ensure numerical columns are float/int
numerical_cols = ['Income', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts',
                  'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
                  'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth', 'Recency']
for col in numerical_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(float)
df[col]
```

Converted numerical columns to float using pd.to_numeric with errors='coerce' to handle invalid entries, filled missing values with 0, ensuring consistent data types for analysis.

```python
# Ensure categorical columns are strings
categorical_cols = ['Education', 'Marital_Status']
for col in categorical_cols:
    df[col] = df[col].astype(str)
df[col]
```

Converted Categorical columns to Str using the method .astype('str') to maintain datatype consistency.

# Step 2: Selecting key numerical variables

```python
key_vars = ['Age', 'Income', 'Total_Spent', 'NumWebPurchases', 'NumCatalogPurchases',
            'NumStorePurchases', 'NumDealsPurchases', 'Recency']
```

# Step 3: Calculate descriptive statistics

```python
from scipy.stats import mode
import numpy as np

stats_dict = {}
for var in key_vars:
    # Calculate mode
    mode_result = mode(df[var])


    if mode_result.mode.size == 1:
        current_mode = mode_result.mode.item()
    elif mode_result.mode.size > 1:
         current_mode = mode_result.mode[0]
    else: # It will handle no mode (e.g., empty or all unique)
        current_mode = np.nan


    stats = {
        'Mean': df[var].mean(),
        'Median': df[var].median(),
        'Mode': current_mode,
        'Variance': df[var].var(),
        'Std': df[var].std(),
        'Min': df[var].min(),
        'Max': df[var].max(),
        'Q1': df[var].quantile(0.25),
        'Q3': df[var].quantile(0.75)
    }

    stats_dict
```

Here, We are caculating the statistical features for the key variables which I have selected( 'Age', 'Income', 'Total_Spent', 'NumWebPurchases', 'NumCatalogPurchases',
            'NumStorePurchases', 'NumDealsPurchases', 'Recency')  and I have stored it in dictionary for easier access.

```python
    # Identify outliers using IQR method

    iqr = stats['Q3'] - stats['Q1']
    lower_bound = stats['Q1'] - 1.5 * iqr
    upper_bound = stats['Q3'] + 1.5 * iqr
    outliers = df[(df[var] < lower_bound) | (df[var] > upper_bound)][var]
    stats['Outlier_Count'] = len(outliers)
    stats['Outlier_Percentage'] = round(len(outliers) / len(df) * 100, 2)
    stats_dict[var] = {k: round(v, 2) if isinstance(v, (int, float)) else v for k, v in stats.items()}

    stats_dict[var]
```

We are Idententifying Outliers from the dictionary which we have already stored the statistical values.

```python
# Create summary table
summary_table = pd.DataFrame(stats_dict).T.reset_index().rename(columns={'index': 'Variable'})
summary_table = summary_table[['Variable', 'Mean', 'Median', 'Mode', 'Variance', 'Std', 'Min', 'Max',
                               'Q1', 'Q3', 'Outlier_Count', 'Outlier_Percentage']]

summary_table
```

Finally we have created a Summary table to clearly distinguish the statistical analysis.

- **Identify variables that could follow specific probability distributions (e.g., Binomial, Normal). Calculate probabilities and expected values based on these distributions.**



**Deliverables**:

# Step 2: Select variables and assign distributions

```python
variables = [
    {'name': 'Age', 'dist': 'Normal', 'params': {'mu': df['Age'].mean(), 'sigma': df['Age'].std()}},
    {'name': 'Income', 'dist': 'Normal (log-transformed)', 'params': {'mu': np.log(df['Income'] + 1).mean(),
                                                                   'sigma': np.log(df['Income'] + 1).std()}},
    {'name': 'NumWebPurchases', 'dist': 'Poisson', 'params': {'lambda': df['NumWebPurchases'].mean()}},
    {'name': 'Response', 'dist': 'Binomial', 'params': {'n': 100, 'p': df['Response'].mean()}}
]

variables
```

We make a list of dictionaries. Each dictionary is like a note card with three pieces of info:

name: The column name (e.g., 'Age').
dist: The distribution type (e.g., 'Normal').
params: A smaller dictionary with numbers that define the distribution.

For each variable:
Age: Calculates the mean (df['Age'].mean()) and standard deviation (df['Age'].std()) of the Age column, calling them mu (average) and sigma (spread).
Income: Takes the natural log of Income + 1 (np.log(df['Income'] + 1)) to make it less skewed, then calculates the mean and standard deviation of the transformed data.

NumWebPurchases: Uses the mean (df['NumWebPurchases'].mean()) as lambda, the average rate of purchases.

Response: Uses the mean (df['Response'].mean()) as p (probability of saying yes) and sets n=100 (number of trials, like asking 100 customers).

# Step 3: Calculate probabilities and expected values

```python
from scipy.stats import norm, binom, poisson

results = []
for var in variables:
    name = var['name']
    dist = var['dist']
    params = var['params']

    if dist == 'Normal':
        mu, sigma = params['mu'], params['sigma']
        expected = mu
        # Example: P(30 <= Age <= 50)
        prob = norm.cdf(50, mu, sigma) - norm.cdf(30, mu, sigma) if name == 'Age' else None
        prob_desc = 'P(30 <= Age <= 50)' if name == 'Age' else 'N/A (log-transformed)'
    elif dist == 'Normal (log-transformed)':
        mu, sigma = params['mu'], params['sigma']
        expected = np.exp(mu + (sigma**2)/2) - 1  # Approx. mean of original scale
        # Example: P(Income > 60000)
        prob = 1 - norm.cdf(np.log(60000 + 1), mu, sigma) if name == 'Income' else None
        prob_desc = 'P(Income > 60000)'
    elif dist == 'Poisson':
        lambda_ = params['lambda']
        expected = lambda_
        # Example: P(NumWebPurchases = 5)
        prob = poisson.pmf(5, lambda_)
        prob_desc = 'P(NumWebPurchases = 5)'
    elif dist == 'Binomial':
        n, p = params['n'], params['p']
        expected = n * p
        # Example: P(at least 20 acceptances in 100 customers)
        prob = 1 - binom.cdf(19, n, p)
        prob_desc = 'P(>= 20 acceptances in 100)'

    results.append({
        'Variable': name,
        'Distribution': dist,
        'Parameters': ', '.join([f"{k}={v:.2f}" for k, v in params.items()]),
        'Expected Value': round(expected, 2),
        'Sample Probability': round(prob, 4) if prob is not None else 'N/A',
        'Probability Description': prob_desc
    })

    results
```

Here, We loop through each variable (Age, Income, etc.) from Step 2 and does two things:
Calculates the expected value (average outcome) for each variable‚Äôs distribution.
Calculates a sample probability (e.g., chance of Age being 30‚Äì50) to show how the distribution works.
It stores the results in a list called results, where each item is a dictionary with details like the variable name, distribution, and calculated values.   For Response, it‚Äôs the average number of ‚Äúyes‚Äù answers in 100 customers (e.g., 15 if 15% say yes).

```python
# Create summary table
summary_table = pd.DataFrame(results)
summary_table
```

# Step 4: Visualize distributions

```python
for var in variables:
    name = var['name']
    dist = var['dist']
    plt.figure(figsize=(8, 6))

    if dist in ['Normal', 'Normal (log-transformed)']:
        sns.histplot(df[name] if dist == 'Normal' else np.log(df[name] + 1), bins=30, kde=True, color='skyblue')
        plt.title(f'{name} Distribution (Fit: {dist})')
        plt.xlabel(name if dist == 'Normal' else f'log({name} + 1)')
    elif dist == 'Poisson':
        sns.histplot(df[name], bins=range(int(max(df[name]) + 1)), kde=False, color='skyblue')
        x = np.arange(0, int(max(df[name]) + 1))
        plt.plot(x, poisson.pmf(x, var['params']['lambda']) * len(df[name]), 'r-', label='Poisson Fit')
        plt.legend()
    elif dist == 'Binomial':
        sns.histplot(df[name], bins=[0, 1, 2], kde=False, color='skyblue')
        plt.title(f'{name} Distribution (Bernoulli, approx. Binomial)')

    plt.ylabel('Frequency')
    plt.savefig(f'/content/{name.lower()}_distribution.png')
    plt.show()
```

Visualized probability distributions for Age, Income, NumWebPurchases, and Response using histograms tailored to their respective distributions (Normal, Log-Normal, Poisson, Binomial). Saved plots as PNG files to illustrate data fit for marketing analysis.

# Step 3: Customer segmentation

```python
# Select features for clustering
cluster_features = ['Age', 'Income', 'Total_Spent', 'NumWebPurchases']
X = df[cluster_features].dropna()
X
```

```python
# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
# Elbow method to find optimal clusters
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

inertia = np.array(inertia)
inertia
```

```python
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.savefig('/content/elbow_plot.png')
plt.show()
```

```python
# Apply K-means with optimal clusters (e.g., k=4, adjust based on elbow plot)
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df.loc[X.index, 'Segment'] = kmeans.fit_predict(X_scaled)
df.head()
```

Selected features (Age, Income, Total_Spent, NumWebPurchases) for K-means clustering to group customers.

Standardized features to ensure fair comparisons, using StandardScaler.
Used the elbow method to find the optimal number of clusters, plotting inertia (elbow_plot.png) to suggest k=4.

Applied K-means with 4 clusters, adding a Segment column to the dataset with cluster labels (0‚Äì3).

Enabled segmentation for profiling distinct customer groups based on behaviors.
